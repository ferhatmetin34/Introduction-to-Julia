## Loops

```julia
n=0

while n<10
n+=1
println(n)
end
n
```

```julia
n=0

while n<10
    n+=1
    println(n)
end
n
```

```julia
letters=["A","B","C","D","E","F","G"]

i=1

while i<= length(letters)
    letter=letters[i]
    println("Hi $letter")
    i+=1
end
```

```julia
for i in 1:10
    println(i)
end
```

```julia
for i in 1:3:10 #step size 3
    println(i)
end
```

```julia
for i in letters
    println(i)
end
```

```julia
m,n=5,5
A=fill(0,(m,n))
```

```julia
for i in 1:m
    for j in 1:n
        A[i,j]=i+j
    end
end
A
```

```julia
B=fill(0,(m,n))

for i in 1:m,j in 1:n
    B[i,j]=i+j
end
B
```
